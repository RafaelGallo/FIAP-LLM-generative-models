```python
# 1.3 - Crie uma função que analise uma e estuture uma resposta
import json

# Criando a função
def fn_analisa_imagem(image_url):
  """ Função que analisa se uma imagem está seguindo ou não as regras de uso da plataforma.
  """

  prompt = [
    {
            "role": "developer",
            "content": [
                {
                  "type": "input_text",
                  "text": """
                      Você é um especialista em segurança de dados e é respossável por fazer análise de imagens e fotos de veículos publicados no marketplace.
                      No marketplace existem algumas regras de publicação e uso, como:
                      - Não é permitido divulgar imagens de veículos com números de celular, telefone, whatsapp, endereço de e-mail, endereço de localização, redes sóciais ou qualquer outra informação que permita a localização e contato que não seja liberado pela plataforma.
                      - Informações com a marca, logotipo, nome do vendedor e nome da concessionária é permitido.

                      Analise a imagem a seguir e valide se ela possui algum tipo de contato que não seja permitido pela plataforma. Responda no formato JSON solicitado e
                      deve conter as tags 'flg_suspeito', 'comentario' e 'descricao'  para cada descrição.
                      **Essa análise deve ser Minuciosa**

                      EXEMPLO DE RESPOSTA:
                      { "flg_suspeito": "suspeito/não suspeito",
                        "comentario": "comentário resumido sobre o motivo da suspeita, se não houver suspeita colcoar 'sem suspeita'.",
                        "descricao": "breve descrição sobre o veículo e um panorama geral da imagem analisada" }

                      Não adicione explicações extras, responda **apenas** com o JSON.
                    """
                  }
              ]
            },
         {
             "role": "user",
             "content": [
                 {
                     "type": "input_text",
                     "text": f"Segue imagem para analise"
                     },
                 {
                     "type": "input_image",
                     "image_url": image_url,
            }
        ],
    }]

  response = client.responses.create(
    model="gpt-4.1-mini",
    temperature=0.7,
    input=prompt
  )

  response_json = json.loads(response.output_text)

  return response_json['flg_suspeito'], response_json['comentario'], response_json['descricao']
```
